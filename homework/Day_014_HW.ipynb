```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

# 忽略警告訊息
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = '../data/'
```

```python
# 讀取檔案
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
age_data = app_train[['HOUSETYPE_MODE', 'AMT_CREDIT']] # subset
age_data['THOUSAND_AMT_CREDIT'] = age_data['AMT_CREDIT'] / 100


# 連續資料離散化
age_data['BINNED_AMT_CREDIT'] = pd.cut(age_data['THOUSAND_AMT_CREDIT'], 10)

print('------------------')
print(age_data['BINNED_AMT_CREDIT'].value_counts())
age_data.head()
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版


"""
Your Code Here
"""
unique_house_type = np.sort(age_data['BINNED_AMT_CREDIT'].unique())

nrows = len(unique_house_type)
ncols = 2
print(nrows)
print(ncols)

plt.figure(figsize=(10,30))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    """
    Your Code Here
    """
    sns.distplot(age_data.loc[(age_data['BINNED_AMT_CREDIT'] == unique_house_type[i]) & \
                              age_data['HOUSETYPE_MODE'], 'THOUSAND_AMT_CREDIT'], 
                 label = "TARGET = 0", hist = False)
    sns.distplot(age_data.loc[(age_data['BINNED_AMT_CREDIT'] == unique_house_type[i]) & \
                              age_data['HOUSETYPE_MODE'], 'THOUSAND_AMT_CREDIT'], 
                 label = "TARGET = 1", hist = False)
    
    plt.title(str(unique_house_type[i]))
plt.show()    
```
